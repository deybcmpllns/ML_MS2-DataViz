```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('customer_combined_cleaned.csv')
df = df[(df['Age'] != 0) & (df['Amount'] != 0)]

sns.set(style='whitegrid')

# Color palettes
gender_palette = {'Male': '#3471eb', 'Female': '#e34889'}
income_palette = sns.color_palette('YlOrBr', as_cmap=False, n_colors=len(df['IncomeLevel'].unique()))
category_palette = sns.color_palette('Set2', n_colors=len(df['ProductCategory'].unique()))
payment_palette = sns.color_palette('dark', n_colors=len(df['PaymentMethod'].unique()))
platform_palette = sns.color_palette('pastel', n_colors=len(df['Platform'].unique()))
interaction_palette = sns.color_palette('hls', n_colors=len(df['InteractionType'].unique()))
sentiment_palette = {
	'Positive': '#43a047',
	'Neutral': '#e5df53',
	'Negative': '#e53935',
	'Very Positive': '#1b5e20',   # dark green
	'Very Negative': '#b71c1c'    # dark red
}

# 1. Distribution of Age
plt.figure(figsize=(7,5))
sns.histplot(df['Age'], bins=20, kde=True, color='#1d3557')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 2. Distribution of Amount
plt.figure(figsize=(7,5))
sns.histplot(df['Amount'], bins=20, kde=True, color='#e63946')
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 3. Gender Count with standout colors
plt.figure(figsize=(5,4))
sns.countplot(x='Gender', data=df, palette=gender_palette)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 4. Income Level Count, warm color ramp
plt.figure(figsize=(6,4))
sns.countplot(x='IncomeLevel', data=df, palette=income_palette, order=df['IncomeLevel'].value_counts().index)
plt.title('Income Level Distribution')
plt.xlabel('Income Level')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 5. Product Category, Set2 for clarity
plt.figure(figsize=(8,5))
sns.countplot(x='ProductCategory', data=df, palette=category_palette, order=df['ProductCategory'].value_counts().index)
plt.title('Product Category Popularity')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 6. Payment Method, dark colors for distinct options
plt.figure(figsize=(7,5))
sns.countplot(x='PaymentMethod', data=df, palette=payment_palette, order=df['PaymentMethod'].value_counts().index)
plt.title('Payment Method Usage')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 7. Platform, pastel palette for visual variety
plt.figure(figsize=(7,5))
sns.countplot(x='Platform', data=df, palette=platform_palette, order=df['Platform'].value_counts().index)
plt.title('Social Platform Usage')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 8. Interaction Type, hls palette for engagement emphasis
plt.figure(figsize=(7,5))
sns.countplot(x='InteractionType', data=df, palette=interaction_palette, order=df['InteractionType'].value_counts().index)
plt.title('Interaction Type Distribution')
plt.xlabel('Interaction Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 9. Sentiment, explicit semantic colors
plt.figure(figsize=(7,5))
sns.countplot(x='Sentiment', data=df, palette=sentiment_palette, order=df['Sentiment'].value_counts().index)
plt.title('Interaction Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 10. Age vs. Amount by Income: color by income, emphasis alpha
plt.figure(figsize=(7,5))
sns.scatterplot(x='Age', y='Amount', hue='IncomeLevel', palette='YlOrBr', data=df, alpha=0.6, edgecolor='k', linewidth=0.5)
plt.title('Age vs. Amount by Income Level')
plt.xlabel('Age')
plt.ylabel('Transaction Amount')
plt.tight_layout()
plt.show()
```
